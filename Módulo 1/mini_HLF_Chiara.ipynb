#### Primeros pasos
Vamos a empezar a programar trozos de código que nos servirán para el HLF:

* Crear un Tablero como una lista de listas de 10x10 que contenga caracteres espacio: " "
* Insertar dos barcos sustituyendo en el Tablero los espacios en blanco por el caracter "B":
    * Uno de 4 casillas en horizontal: casillas [1][0], [1][1], [1][2], [1][3]
    * Uno de 3 casillas en vertical:   casillas [3][3], [4][3], [5][3]
* Escribir una celda de código separada de las anteriores para simular que el usuario está efectuando un disparo: 
    *  Solicite al usuario dos coordenadas `i,j` mediante dos inputs
    *  Compruebe si esas coordenadas están ocupada por una casilla barco (que haya una en "B")
        * Si hay una "B":
            * Cambiar el contenido de esa posición por una "x"
            * Imprimir por pantalla el texto "Tocado en posición `i,j`"
        * Si hay un " ":
            * Cambiar el contenido de esa posición por una "o"
            * Imprimir por pantalla el texto "Agua"
    * Imprimir por pantalla el Tablero
* Para comprobar que funciona bien la celda de disparar ejecutarla varias veces metiendo distintas coordenadas

### Bonus

* Resuélvelo con funciones:
    * Una para solicitar una coordenada válida a la que llamarás dos veces, una par i y otra par j
    * Una `disparar` que devuelva `True` si el usuario ha acertado o `False` si no.
    


```python
def crear_tablero():
    """ Función para crear un nuevo tablero de 10x10. Devuelve el tablero; se puede guardar en una nueva variable. """
    tablero = [[" " for fila in range(10)] for i in range(10)]
    # Fila es una lista completa del tablero
    # i es el contenido de cada celda (la "columna" de la fila)
    return tablero
```

```python
def imprimir_tablero(tablero):
    """ Función para imprimir un determinado tablero. """
    for fila in tablero:
        for i in fila:
            print(f"[{i}]",end=" ")
        print("")
```

```python
# 1. Insertar dos barcos sustituyendo en el tablero los espacios en blanco por el caracter "B":
mi_tablero = crear_tablero()

# Uno de 4 casillas en horizontal: casillas [1][0], [1][1], [1][2], [1][3]
mi_tablero[1][0] = "B"
mi_tablero[1][1] = "B"
mi_tablero[1][2] = "B"
mi_tablero[1][3] = "B"

# Uno de 3 casillas en vertical: casillas [3][3], [4][3], [5][3]
mi_tablero[3][3] = "B"
mi_tablero[4][3] = "B"
mi_tablero[5][3] = "B"

imprimir_tablero(mi_tablero)
```

```python
# 2. Simular que el usuario está efectuando un disparo:
# Input con las coordenadas del disparo
input_fila = int(input("Introduce la coordenada i:"))
input_columna = int(input("Introduce la coordenada j:"))

# 3. Comprobar si esas coordenadas están ocupadas por una casilla barco ("B")
tablero_vacio = crear_tablero() # Tengo que crear un tablero vacío, si no se imprime el tablero con las "B"
# Si hay una "B"
if mi_tablero[input_fila][input_columna] == "B":
    # Cambiar el contenido de esa posición por una "X"
    tablero_vacio[input_fila][input_columna] = "X"    
    # Imprimir por pantalla el texto "Tocado en posición `i,j`"
    print(f"Tocado en posición {input_fila}, {input_columna}.")
#Si hay un " ":
if mi_tablero[input_fila][input_columna] == " ":        
    # Cambiar el contenido de esa posición por una "o"
    tablero_vacio[input_fila][input_columna] = "O"    
    # Imprimir por pantalla el texto "Agua"
    print("Agua.")

# 4. Imprimir por pantalla el Tablero
imprimir_tablero(tablero_vacio)

# NOTA: en un tablero 10:10, la última celda es la coordenada 9:9, porque Python cuenta 0:0
```

```python
# 5. En la celda de disparar, si el usuario acierta a tocar un barco, vuelve a disparar:
tablero_vacio = crear_tablero()

while any("B" in fila for fila in mi_tablero): # Como es lista de listas, poniendo "while "B""" asume que toda la fila tiene que ser B; de esta forma, busca dentro de cada lista
    input_fila = int(input("Introduce la coordenada i:"))
    input_columna = int(input("Introduce la coordenada j:"))

    if mi_tablero[input_fila][input_columna] == "B":
        tablero_vacio[input_fila][input_columna] = "X"    
        print(f"Tocado en posición {input_fila}, {input_columna}.")
        imprimir_tablero(tablero_vacio)

    if mi_tablero[input_fila][input_columna] == " ":        
        tablero_vacio[input_fila][input_columna] = "O"    
        print("Agua.")
        imprimir_tablero(tablero_vacio)
        print("Turno de tu contrincante.")
        break
```
