## Ejercicio 2
Eleva todos los elementos de la lista al cuadrado
> ```ejer_2 = [1,2,3,4,5]```

````python
#MANERA 1: hacemos un bucle "for" para que haga la operación con todos los elementos de la lista
ejer_2 = [1,2,3,4,5]
for i in range(len(ejer_2)): #donde "i" es la posición de cada elemento de la lista y el bucle va a coger la longitud del rango de la lista "ejer_2"
    ejer_2[i] **= 2 #por cada posición de la lista, se hace el karratu
ejer_2 
````

````python
ejer_2 = [1,2,3,4,5]
cuadrados = [x**2 for x in ejer_2] #donde "x" es la posición de los componentes de la lista y se le hace el cuadrado (el bucle "for" de otra manera)
print(cuadrados)
````

## Ejercicio 3
Crea una lista nueva con todas las combinaciones de las siguientes dos listas:
> ```ejer_3_1 = ["Hola", "amigo"]```
>
> ```ejer_3_2 = ["Que", "tal"]```

Obten el siguiente output:

```['Hola Que', 'Hola tal', 'amigo Que', 'amigo tal']```

````python
ejer_3_1 = ["Hola", "amigo"]
ejer_3_2 = ["Que", "tal"]
ejer_3 = [] #se crea una lista vacía a la que iremos añadiendo todos los elementos string de las listas
for i in ejer_3_1: #"i" son los elementos de la lista ejer_3_1 y "j" de la lista ejer_3_1. El bucle "for" lo que hace es recorrer todos los elementos para crear la nueva lista
    for j in ejer_3_2:
        ejer_3.append(i + " " + j) #el comando "append" lo añade al final de la lista. Como está vacía al principio, los elementos añadidos serán los únicos elementos
ejer_3
````

````python
tupla_1 = (1,2,3)
tupla_2 = (4,) #cuando la tupla tiene 1 elemento, hay que ponerle la "," después del elemento para que se sepa que es una tupla
print(type(tupla_1))
print(type(tupla_2))
````

````python
new_tupla = (2, 1.2, "hello", False)
print(new_tupla[0])
print(new_tupla[len(new_tupla) - 1]) #la posición -1 (última) de la longitud de la tupla
````

````python
#como a las tuplas no se les puede añadir elementos ni hacer modificaciones como a las listas, hay que transformarlo en una lista y después de nuevo pasarlo a tupla:
tupla_list = list(new_tupla)
print(tupla_list)
tupla_list.append("bitches") #el comando "append" lo añadirá al final de la lista
print(tupla_list)
newest_tupla = tuple(tupla_list)
newest_tupla
````

````python
#como en la celda anterior, tenemos que partir de una lista porque las tuplas no se pueden modificar. La versión de la tupla en lista es la "tupla_list":
posición = len(tupla_list) // 2 #la posición a eliminar será el cociente de la división de la longitud de la lista / 2
print(posición) #posición que se quiere eliminar
tupla_list.pop(posición) #para quitar el elemento de la posición media
print(tupla_list)
final_tupla = tuple(tupla_list) #lo volvemos a convertir en tupla
final_tupla
````
