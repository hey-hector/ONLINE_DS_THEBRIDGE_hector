## Ejercicio 6
Mismo programa que en 5, pero en esta ocasión tienen que ser tres inputs y dos salidas. Una de las salidas que nos indique si todos son iguales, y la otra si al menos dos inputs sí que lo son

````python
a = input("a")
b = input("b")
c = input("c")
if a == b == c:
    print(f"{a}, {b} y {c} son iguales")
elif a == b and a != c and b != c:
    print(f"{a} y {b} son iguales pero {c} es diferente")
elif a == c and b != c and a != b:
    print(f"{a} y {c} son iguales pero {b} es diferente")
elif b == c and a != b and a != c:
    print(f"{b} y {c} son iguales pero {a} es diferente")
else:
    print(f"{a}, {b} y {c} son diferentes")
````

## Ejercicio 7
Crea un programa que recoja dos inputs. Tiene que comprobar si su suma es igual, superior o inferior a 10, cada condición por separado

````python
número_1 = input("Número 1")
número_2 = input("Número 2")
número_1_deci = float(número_1)
número_2_deci = float(número_2)
if número_1_deci + número_2_deci == 10:
    print("La suma de sus números es igual a 10")
elif número_1_deci + número_2_deci > 10:
    print("La suma de sus números es mayor a 10")
else:
    print("La suma de sus números es menor a 10")
````

## Ejercicio 8
Razona sin ejecutar código el output que obtendremos de las siguientes sentencias
1. True and True and False
2. not ((True or False) and (True or False))
3. (False or False or False or True) and False
4. not False and True and not True

Compruébalo después con python

1. False
2. not (True) and (True) = False and True = False
3. True and False = False
4. True and True and False = False

````python
'''
1. No puede ser True y False a la vez, por lo que el output sería False
2. Al estar la condición or, si hay un True el output ya es True, por lo que:
  Primera parte -> not(True or False): not(True): False
  Segunda parte -> True or False: True
  Combinadas -> False and True
  Respuesta final: False
3. Primera parte -> al estar la condición or, como hay un True el output ya es True
   Combinadas -> True and False
   Respuesta final: False
4. True and True and False: al no poder ser True y False a la vez, el output será False

COMPROBACIÓN:
'''
print(True and True and False)
print(not ((True or False) and (True or False)))
print((False or False or False or True) and False)
print(not False and True and not True)
````

## Ejercicio 10
Para el siguiente string se pide imprimir por pantalla los siguientes casos

    "Lo primero de todo, cómo están los máquinas"

1. Pasarlo todo a mayúsculas
2. Pasarlo todo a minúsculas
3. Pasar a mayúsculas solo las iniciales de las palabras
4. Crea una lista dividiendolo por sus espacios
