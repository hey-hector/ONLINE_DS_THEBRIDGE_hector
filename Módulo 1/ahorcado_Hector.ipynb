````python
#Vamos a crear clases y objetos para este ejercicio porque hay muchas cosas para tener en cuenta y prefiero hacerlo uno a uno:

class ahorcado:
    #zona de atributos propios que no cambian. Que yo sepa, todo va a ir cambiando, así que lo dejo vacío (por ahora)
    def __init__(self, turno = True, vidas:int = 6): #como siempre será mi turno, le ponemos la condición booleana 'True'
        self.vidas = vidas
        self.turno = turno
    def juego(self): #al tener un input aquí para declarar la letra, no hace falta ponerla como atributo para la función/método
        letras = input("¿Qué letra quieres usar?")
        letras = str(letras)
        letras = letras.upper()
        return letras
    def acierto(self, letras): #aquí sí hay que ponerlo porque sino esta función/método "acierto" no va a saber lo que es el atributo "letras"
        if letras == "M" or letras == "A" or letras == "T" or letras == "D":
            #self.vidas = self.vidas (no es necesario ponerlo porque el Nº de vidas se queda igual)
            return "Es tu día de suerte"
        else:
            self.vidas = self.vidas - 1
            return f"Fallaste. Te quedan {self.vidas} vidas"
    def palabra_final(self): #Esto sólo vale para la primera letra. Tendría que hacer un bucle para cada letra e ir rellenándolo
        palabra_correcta = ["M", "A", "T", "A", "D", "A"] #no me sale
        if letras == palabra_correcta[0]:
            return "M _ _ _ _ _"
        elif letras == palabra_correcta[1]:
            return "_ A _ A _ A"
        elif letras == palabra_correcta[2]:
            return "_ _ T _ _ _"
        elif letras == palabra_correcta[4]:
            return "_ _ _ _ D _"
        else:
            return "Try Again Darling"
````
