### 1. Utiliza map para elevar al cuadrado cada número en la lista [1, 2, 3, 4, 5].

Solución esperada: [1, 4, 9, 16, 25]

```python
lista = [1,2,3,4,5]
lista = map(lambda x: x**2 , lista)
print(list(lista))
```

### 2. Filtrar números pares de una lista:

Emplea filter para obtener sólo los números pares de la lista [1, 2, 3, 4, 5, 6].

Solución esperada: [2, 4, 6]

```python
lista = [1,2,3,4,5,6]
nums_pares = filter(lambda x: x % 2 == 0, lista)
print(list(nums_pares))
```

### 3. Convertir grados Celsius a Fahrenheit:

Usa map y una función lambda para convertir una lista de temperaturas en grados Celsius [0, 10, 20, 30] a Fahrenheit.

Solución esperada: [32.0, 50.0, 68.0, 86.0]

```python
celsius = [0,10,20,30]
fahrenheit = map(lambda x: (x * 9/5) + 32, celsius)
print(list(fahrenheit))
```

### 4. Sumar los cuadrados de los números pares de una lista:

Combina filter, map y reduce (importa reduce desde functools) para sumar los cuadrados de los números pares en [1, 2, 3, 4, 5, 6].

Solución esperada: 56

```python
from functools import reduce
```

```python
lista = [1,2,3,4,5,6]
```

```python

nums_pares = list(filter(lambda x: x % 2 == 0, lista)) #ponemos list antes de la función filter porque queremos que el tipo del output (nums_pares) sea una lista
print((nums_pares))
print(type(nums_pares))
```

```python
cuadrados = list(map(lambda x: x**2, nums_pares))
print(cuadrados)
```

```python
resultado = reduce(lambda x,y: x + y, cuadrados)
print(resultado)
```

### 5. Encadenar map, filter, y reduce para resolver un problema complejo:

Dada una lista de strings que representan números enteros y "NA" (por ejemplo, ["2", "5", "NA", "9", "10"]), convierte los números a enteros, filtra los "NA", suma el cuadrado de los números que son mayores que 5 y menores que 10.

Solución esperada: Un único valor que es la suma de los cuadrados de los números válidos.

```python
from functools import reduce
```

```python
lista = ["2", "5", "NA", "9", "10"]
filtrado = list(filter(lambda x: x != "NA", lista))
print(list(filtrado))
```

```python
nums_enteros = list(map(lambda x: int(x), filtrado))
print(nums_enteros)
```

```python
"""
Interpretación 1: hacer el cuadrado de los números de la lista que sean menores de 5 y mayores de 10. Es decir, sólo lo hará del 9.
reduce(lambda resultado,x: operación_con_resultado + (x if(condición_1_de_x) else condición_2_de_x), sobre qué función/variable se quiere hacer, 0 [empieza de valor 0])

Otra manera para hacer esta celda en 2 pasos: 
nombre1 = filter(lambda x: 5 < x < 10, nums_enteros)
nombre2 = map(lambda x: x**2, nombre1)
"""
suma_cuadrados = reduce(lambda resultado,x: resultado**2 + (x if(5 < x < 10) else 0), nums_enteros, 0) 
print(suma_cuadrados)
```

```python
"""
Interpretación 2: hacer el cuadrado de todos y después coger los valores que sean menores de 5 y mayores de 10 de esa nueva lista. Es decir, ninguno; por lo que el
resultado será 0 (condición del 'else')
"""
suma = list(map(lambda x: x**2, nums_enteros))
print(suma)
```

```python
suma_cuadrados_dos = reduce(lambda resultado,x: resultado + (x if(5 < x < 10) else 0), suma, 0) #
print(suma_cuadrados_dos)
```
