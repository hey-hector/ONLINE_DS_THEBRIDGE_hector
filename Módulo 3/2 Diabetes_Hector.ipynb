## Building a regression model

The data have already been split to predictor (*data*) and response (*target*) variables. Given this information, we'll apply what we have previously learned about linear regression and apply the algorithm to the diabetes dataset.

#### In the cell below, import the `linear_model` class from `sklearn`. 

```python
#Ya tenemos los datos inspeccionados y listos. Al ser un problema donde buscamos predecir un valor numérico continuo (la progresión de la enfermedad), 
#la regresión lineal es el punto de partida ideal.
#Código para importar la clase necesaria e inicializar el modelo:
```

```python
from sklearn import linear_model
# 1. Crear una instancia del modelo de Regresión Lineal
modelo_diabetes = linear_model.LinearRegression()

# 2. Entrenar el modelo usando todos los atributos (X) y el objetivo (y)
# Nota: Aquí usamos 'diabetes.data' (X) y 'diabetes.target' (y)
modelo_diabetes.fit(diabetes.data, diabetes.target)
# Al ejecutar .fit(), el algoritmo ha calculado los coeficientes (pesos) para cada uno de los 10 atributos (edad, bmi, s1, etc.) y el intercepto, 
# buscando la línea (o hiperplano en este caso) que minimice el error cuadrático respecto a los 442 pacientes.
```

#### Create a new instance of the linear regression model and assign the new instance to the variable `diabetes_model`.

```python
#Vamos a inicializar el modelo siguiendo la nomenclatura. Al asignar la instancia a diabetes_model, ya quedará lista para recibir los datos de entrenamiento.

# Crear la instancia del modelo de regresión lineal
diabetes_model = linear_model.LinearRegression()

#Con este paso, ya tenemos el objeto configurado con los algoritmos de mínimos cuadrados listos para ser aplicados.
```

#### Next, let's split the training and test data.

Define `diabetes_data_train`, `diabetes_target_train`, `diabetes_data_test`, and `diabetes_target_test`. Use the last 20 records for the test data and the rest for the training data.

```python
# Para dividir el conjunto de datos de forma manual según tu instrucción (los últimos 20 para prueba y el resto para entrenamiento), utilizaremos el slicing (segmentación)
# de arreglos de NumPy.
# Aquí tienes la definición de las variables:

# Definición de los datos de entrenamiento (todos excepto los últimos 20)
diabetes_data_train = diabetes.data[:-20]
diabetes_target_train = diabetes.target[:-20]

# Definición de los datos de prueba (los últimos 20 registros)
diabetes_data_test = diabetes.data[-20:]
diabetes_target_test = diabetes.target[-20:]
# Verificación de las dimensiones
print(f"Registros para entrenamiento: {diabetes_data_train.shape[0]}")
print(f"Registros para prueba: {diabetes_data_test.shape[0]}")

# ¿Qué acabamos de hacer?
# [:-20]: Toma desde el inicio del arreglo hasta el registro número 422 (excluyendo los últimos 20).
# [-20:]: Toma desde el registro 422 hasta el final (los últimos 20 registros).
# Esta separación es fundamental para evaluar si nuestro modelo realmente ha aprendido a generalizar o si solo ha memorizado los datos de entrenamiento.
```

Perform a brief exploratory analysis to investigate the conditions necessary for linear regression. This analysis will help you understand the distribution of the target variable and its correlations with other features.

Fit the training data and target to `diabetes_model`. Print the *intercept* and *coefficients* of the model.

```python
# Ahora que tenemos los datos divididos, vamos a entrenar el modelo utilizando exclusivamente el conjunto de entrenamiento. Una vez ajustado, extraeremos los parámetros 
# matemáticos que el modelo ha aprendido.
# Aquí tienes el código para realizar el ajuste y mostrar los resultados:

# 1. Entrenar (ajustar) el modelo con los datos de entrenamiento
diabetes_model.fit(diabetes_data_train, diabetes_target_train)

# 2. Obtener y mostrar el intercepto y los coeficientes
print(f"Intercepto (b0): {diabetes_model.intercept_:.4f}")
print("\nCoeficientes (uno por cada atributo):")
for nombre, coef in zip(diabetes.feature_names, diabetes_model.coef_):
    print(f"{nombre}: {coef:.4f}")

# ¿Qué significan estos valores?
# Intercepto: Es el valor base de la progresión de la enfermedad cuando todos los atributos (edad, bmi, etc.) son cero.
# Coeficientes: Indican cuánto cambia la progresión de la enfermedad por cada unidad que aumenta un atributo.
# Un coeficiente positivo (como en bmi o s5) indica que a mayor valor, mayor progresión de la diabetes.
# Un coeficiente negativo indica una relación inversa.
# Dado que los datos están normalizados, puedes comparar los coeficientes directamente: aquel con el valor absoluto más alto es el que tiene mayor importancia en la predicción.
```

#### Inspecting the results

From the outputs you should have seen:

- The intercept is a float number.
- The coefficients are an array containing 10 float numbers.

This is the linear regression model fitted to your training dataset.

#### Using your fitted linear regression model, predict the *y* of `diabetes_data_test`.

```python
# Ahora que el modelo ha "aprendido" la relación matemática a través del intercepto y los 10 coeficientes, podemos ponerlo a prueba.
# Usaremos el método .predict() pasándole únicamente las características de los últimos 20 pacientes (diabetes_data_test). El modelo aplicará la fórmula lineal y nos devolverá 
# sus estimaciones:

# Realizar las predicciones usando los datos de prueba
diabetes_y_pred = diabetes_model.predict(diabetes_data_test)

# Mostrar las primeras 5 predicciones comparadas con los valores reales
print("Primeras 5 predicciones vs. Valores reales:")
for i in range(5):
    print(f"Predicción: {diabetes_y_pred[i]:.2f} | Real: {diabetes_target_test[i]}")

# ¿Qué representa diabetes_y_pred?
# Es un arreglo que contiene los valores de progresión de la enfermedad que el modelo estima para esos 20 pacientes basándose en su edad, IMC, presión arterial y niveles de suero.
```

#### Print your `diabetes_target_test` and compare with the prediction. 

```python
# Para comparar directamente los resultados, imprimiremos ambos arreglos. Esto te permitirá observar visualmente la diferencia (el residuo) entre lo que el modelo predijo 
# y lo que realmente sucedió con esos 20 pacientes.
# Aquí tienes el código para mostrar la comparativa:

# Imprimir los valores reales y las predicciones
print("Valores Reales (Target Test):")
print(diabetes_target_test)

print("\nPredicciones del Modelo:")
print(diabetes_y_pred)

#Una comparativa más legible fila por fila:
import pandas as pd

comparativa = pd.DataFrame({'Real': diabetes_target_test, 'Predicción': diabetes_y_pred.round(2), 'Error (Diferencia)': (diabetes_target_test - diabetes_y_pred).round(2)})

print("\nTabla Comparativa de los últimos 20 registros:")
print(comparativa)

# ¿Qué observar en esta comparación?
# Cercanía: Verás que en algunos casos el modelo se acerca bastante (errores pequeños), mientras que en otros la diferencia es significativa.
# Tendencia: Observa si el modelo tiende a sobreestimar (predicción mayor al real) o subestimar (predicción menor al real) los valores.
# Variabilidad: Como los valores de progresión van aproximadamente de 25 a 340, un error de 10 o 20 unidades se considera aceptable, pero errores de más de 50 indican que
# el modelo no capturó bien la complejidad de ese paciente específico.
```

#### Is `diabetes_target_test` exactly the same as the model prediction? Use different error metrics

```python
# No, los valores de diabetes_target_test no son exactamente iguales a las predicciones del modelo. En un problema de regresión con datos del mundo real, es prácticamente
# imposible que coincidan a la perfección, ya que la relación entre las variables no es puramente lineal y existen factores no medidos.
# Para cuantificar qué tan lejos están las predicciones de la realidad, calcularemos las métricas de error que utilizamos anteriormente. Aquí tienes el código para obtener
# MAE, MSE, RMSE y R²:

from sklearn import metrics
import numpy as np

# 1. MAE: Error Absoluto Medio
mae = metrics.mean_absolute_error(diabetes_target_test, diabetes_y_pred)

# 2. MSE: Error Cuadrático Medio
mse = metrics.mean_squared_error(diabetes_target_test, diabetes_y_pred)

# 3. RMSE: Raíz del Error Cuadrático Medio
rmse = np.sqrt(mse)

# 4. R²: Coeficiente de Determinación
r2 = metrics.r2_score(diabetes_target_test, diabetes_y_pred)

print(f"Métricas de Evaluación para el Set de Prueba:")
print(f"-------------------------------------------")
print(f"MAE:  {mae:.2f} (Error promedio en unidades de progresión)")
print(f"MSE:  {mse:.2f}")
print(f"RMSE: {rmse:.2f} (Desviación típica del error)")
print(f"R²:   {r2:.4f} (Proporción de varianza explicada)")

# ¿Cómo interpretar estos números?
# Diferencia Visual: Si el modelo fuera perfecto, el R² sería 1.0 y todas las métricas de error serían 0.
# MAE / RMSE: Estos valores te dicen cuántas unidades de "progresión de enfermedad" suele fallar el modelo. Por ejemplo, si el MAE es de 40, significa que, en promedio,
# tus predicciones se alejan 40 puntos del valor real.
# R² (Bondad de ajuste): En el set de prueba, este valor nos indica qué tan bien generaliza el modelo con datos nuevos. Un valor cercano a 0.5 suele ser común en este dataset,
# lo que significa que el modelo explica el 50% de lo que le ocurre al paciente.
```

#### Iterate with a new model to assess if any improvements can be made compared to the previous one. Consider reducing the number of variables or using `StandardScaler` to enhance model performance.

```python
# the number of variables or using `StandardScaler` to enhance model performance.
# Para mejorar el modelo, vamos a implementar dos estrategias clave: la estandarización de las variables y la selección de características.
# Aunque los datos originales de diabetes ya están centrados, el uso de StandardScaler es una mejor práctica que asegura que todas las variables tengan media 0 y varianza 1,
# facilitando el trabajo del algoritmo. Además, eliminaremos las variables con menor impacto para reducir el "ruido".
# Aquí tienes el código para este nuevo modelo:

from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import pandas as pd

# 1. Selección de variables (Feature Selection)
# Estudios previos sugieren que BMI, S5 (triglicéridos) y BP son los predictores más fuertes
features_reducidas = ['bmi', 's5', 'bp', 's3', 's6']
X_train_df = pd.DataFrame(diabetes_data_train, columns=diabetes.feature_names)[features_reducidas]
X_test_df = pd.DataFrame(diabetes_data_test, columns=diabetes.feature_names)[features_reducidas]

# 2. Estandarización
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_df)
X_test_scaled = scaler.transform(X_test_df)

# 3. Entrenamiento del nuevo modelo
nuevo_modelo_diabetes = LinearRegression()
nuevo_modelo_diabetes.fit(X_train_scaled, diabetes_target_train)

# 4. Predicción y Evaluación
y_pred_nuevo = nuevo_modelo_diabetes.predict(X_test_scaled)
mse_nuevo = mean_squared_error(diabetes_target_test, y_pred_nuevo)
r2_nuevo = r2_score(diabetes_target_test, y_pred_nuevo)

print(f"--- Comparativa de Modelos ---")
print(f"Modelo Anterior -> MSE: {metrics.mean_squared_error(diabetes_target_test, diabetes_y_pred):.2f}, R2: {metrics.r2_score(diabetes_target_test, diabetes_y_pred):.4f}")
print(f"Modelo Nuevo    -> MSE: {mse_nuevo:.2f}, R2: {r2_nuevo:.4f}")

# ¿Qué mejoras estamos buscando?
# Reducción de Varianza: Al usar menos variables (solo las 5 más importantes), el modelo es menos propenso a ajustarse a ruidos aleatorios de los datos de entrenamiento 
# (overfitting).
# Interpretabilidad: Un modelo con 5 variables es mucho más fácil de explicar clínicamente que uno con 10.
# Estabilidad: StandardScaler asegura que si en el futuro ingresas datos en escalas diferentes, el modelo mantenga su consistencia.
# Nota sobre los resultados: En este dataset específico, verás que el R² puede no subir drásticamente (o incluso bajar levemente) porque el volumen de datos es pequeño 
# (442 registros). Sin embargo, el modelo nuevo suele ser más robusto para datos futuros.
```
