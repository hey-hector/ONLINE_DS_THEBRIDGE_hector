## Ejercicio 10
El amargor de las cervezas se mide por su nivel de IBUs.
1. ¿Cuál es el amargor medio de las cervezas?
2. ¿El máximo? ¿Y el mínimo?
3. ¿Qué cerveza es la que tiene el amargor máximo? ¿Cuál es la de amargor mínimo?

```python
cerves["ibu"].mean()
```

```python
cerves["ibu"].max()
```

```python
cerves["ibu"].min()
```

```python
cerves.loc[cerves['ibu'].idxmax(), 'name'] #identifica el máximo de la columna deseada (ibu) y devuelve el valor de la columna que necesitas (name)
#sólo identifica el primero
```

```python
cerves[["name", "ibu"]][cerves["ibu"] == cerves["ibu"].max()] #identifica todos los ibu con valores máximos del dataframe
```

```python
cerves.loc[cerves['ibu'].idxmin(), 'name'] #primer valor mínimo
```

```python
cerves[["name", "ibu"]][cerves["ibu"] == cerves["ibu"].min()] #identifica todos los ibu con valores mínimos del dataframe
```

## Ejercicio 17

## Ejercicio 18
Junta los dos datasets que tienes

```python
#ambos datasets (cerves y cervecerías) tienen columnas diferentes y dos iguales ("Unnamed: 0" y "name"). Lo que pasa es que se refieren a cosas diferentes: uno
#habla de ids y nombres de cerveza y el otro de cervecerías, por lo que usaremos el dataset del ejercicio anterior (dataset_2) con los nombres cambiados ya para que
# cada columna indique una cosa diferente y no haya cruzamiento de datos. Usamos el comando "merge"
new_dataset = pd.concat([cerves, dataset_2], axis=1) #"axis=1" me dice que vamos a concatenar los datasets por columnas, es decir, las columnas del segundo dataset se añaden
#después (a la derecha) del primer dataset
new_dataset
```
