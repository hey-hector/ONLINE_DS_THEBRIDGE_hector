````python
#PASO 1: Obtener el diccionario de géneros para traducir los IDs a nombres:
gen_url = f"{url_base}/genre/movie/list"
params_gen = {"api_key": api_key, "language": "es-ES"}
genres_data = requests.get(gen_url, params=params_gen).json()
genre_dict = {g['id']: g['name'] for g in genres_data['genres']}
````

````python
#PASO 2: Buscar películas que contengan una palabra que empiece por "H":
search_url = f"{url_base}/search/movie"
params = {"api_key": api_key, "query": mi_inicial, "language": "es-ES", "page": 1}
data = requests.get(search_url, params=params).json()
````

````python
#PASO 3: Procesar y estructurar la información:
lista_peliculas = [] #creamos una lista vacía para que con un bucle la vaya rellenando

for movie in data.get('results', []):
    # Convertir IDs de géneros a nombres legibles:
    nombres_generos = ", ".join([genre_dict.get(gi, "Desconocido") for gi in movie.get('genre_ids', [])])
    
    lista_peliculas.append({"id": movie.get('id'), "title": movie.get('title'), "release_date": movie.get('release_date'), "genres": nombres_generos,
                            "vote_average": movie.get('vote_average'), "overview": movie.get('overview')})
````

````python
#PASO 4: Crear DataFrame y guardar en .csv:
df_peliculas = pd.DataFrame(lista_peliculas)
df_peliculas.to_csv("peliculas_H.csv", index=False, encoding='utf-8')

print("Archivo 'peliculas_H.csv' generado con éxito.")
print(df_peliculas[['title', 'genres']].head()) #sólo salen las 5 primeras
````
