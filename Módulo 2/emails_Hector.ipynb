# Extrae los dominios de los correos con una expresión regular

```python
import re

emails = '''
miguellc@thebridgeschool.tech
miguel.cruz@uc3m.es
miguel-12-2000@mi-correo.net
miguellc@s@thebridgeschool.es
'''
```

Ejemplo:
* @thebridgeschool.tech
* @uc3m.es
* @mi-correo.net
* @thebridgeschool.es

```python
# La regex busca la última '@' y captura lo que sigue (letras, números, puntos y guiones)
# El '$' asegura que llegue hasta el final de la línea en modo multilínea:
re.findall(r'@([\w.-]+\.[a-z]{2,})$', emails, re.MULTILINE)
```
